We needed to expand our testing of this facet of "abstraction", so:

* Need to know how to return multiple values, and
* Need to know how to "unwrap" the return values from a function (see example below)

````python
from random import randrange

# declare a function that returns multiple values
def get_3_ducks():
    all_ducks = [
        'Huey','Dewey','Louie','Donald',
        'Daisy','Scrooge','Daffy',
        'Darkwing','Howard','Disco'
        ]
    num_ducks = len(all_ducks)
    duck1 = all_ducks[randrange(num_ducks)]
    duck2 = all_ducks[randrange(num_ducks)]
    duck3 = all_ducks[randrange(num_ducks)]
    return duck1, duck2, duck3


# "unwrap" the multiple returned values from a function call
d1, d2, d3 = get_3_ducks()
print(f"{d1}, {d2}, {d3}")
d1, d2, d3 = get_3_ducks()
print(f"{d1}, {d2}, {d3}")
````

* Regular Expressions
    - You will need to recognize and interpret a regular expression...you will NOT need to write one from scratch

        - Character classes
            - \d  == numeric digit >> [0-9]
            - \w  == "word" character >> [A-Za-z0-9]
            - \s  == "space" character
            - \S  == all but "space" character
        - Custom character class (note that if you are viewing this as text, there should ONLY be a square bracket...the backslash is there to "escape" it for the markdown language that this is written in)
            - \[abc\] finds a or b or c
            - \[Yy\] finds Y or y

            - \[A-F\] finds A,B,C,D,E,F

        - Quantifiers
            - b{3} find exactly 3 b's ("bbb")
            - x{6} find exactly 6 x's ("xxxxxx")
            - v{2,5} finds 2, 3, 4 or 5 v's  ("vv", "vvv", "vvvv", "vvvvv")
            - x{,6} find up to 6 x's
            - x{6,} find at least 6 x's
            - x? - find 0 or 1 x ==> {,1}
            - x* - find 0 or many x's 
            - x+ - find 1 or many x's ==> {1,}

## Dictionaries

* What are they?
* How are they different from a list?
    - implied order?
    - nature of "index"
* key vs value
    - keys must be unique
    - values can be duplicated
* Basic operations:
    - Initialize, e.g.,:
      ```
      empty_dict = {}
      person = {
        "first_name": "Paul",
        "last_name": "York"
      }
      ```
    * Access a value (use "key" as index)
      ```
      print(person['first_name'])  # "Paul"
      ```
    * Add a value (just reference a new key)
      ```
      person['age'] = 29
      ```
    * Change a value (just assign a new value)
      ```
      person['age'] = 47
      ```
    * Remove a value (remove a single key/value pair)
      ```
      del person['age']
      ```
    * Iteratate over a dictionary
      ```
      # HOW WE DID THIS WITH A LIST USING INDEXES
      for idx in range(len(lst)):
          val = lst[idx]
          print(val)

      # WITH LISTS, YOU CAN ALSO JUST ITERATE OVER VALUES
      # (functionally equivalent to the above)
      for val in lst:
          print(val)
      ```

````python
easter_stash = {
    'Cadbury Eggs': 5,
    'Jelly Beans': 55,
    'Chocolate Bunnies': 2,
}

# WITH DICTIONARIES, YOU ITERATE OVER THE KEYS (by default)
for item in easter_stash:
    # can use both key AND value
    print(item + ': ', end='')  # "item" is the key
    num = easter_stash[item]    # Access the value using the key
    print(num)
````

## Python File Management

* Python os, shutil and sys modules (what is what?)
* Python Equivalents to Command Line (CMD==Windows and SH==Mac/Linux)
    - Won't need to "write" these, but you should be able to recognize them
    - View Present/Current Working Directory
        - CMD: cd
        - SH: pwd
        - Python: os.getcwd()
    - Change Directory
        - CMD: cd {path}
        - SH: cd {path}
        - Python: os.chdir()
    - List Directory Contents
        - CMD: dir {path} (path can be left out for "current dir")
        - SH: ls {path}   (path can be left out for "current dir")
        - Python: os.listdir() # for single directory
        - Python: os.walk()    # for whole directory tree
    - Copy Files
        - CMD: copy {file} {newfile}
        - SH: cp {file} {newfile}
        - Python: shutil.copyfile()
    - Copy Whole Directory
        - CMD: xcopy /s {dir} {newdir}
        - SH: cp -r {dir} {newdir}
        - Python: shutil.copytree()
    - Move or Rename File
        - CMD: move {file} {newfile}
        - SH: mv {file} {newfile}
        - Python: shutil.move()
    - Delete File
        - CMD: del {file}
        - SH: rm {file}
        - Python: os.remove()
    - Make Directory
        - CMD: mkdir {newdir}
        - SH: mkdir {newdir}
        - Python: os.mkdir()
    - Delete Directory
        - CMD: rmdir {dir} (add /s is the directory is not empty)
        - SH: rm {dir} (add -r if the diretory is not empty)
        - Python: shutil.rmtree()
